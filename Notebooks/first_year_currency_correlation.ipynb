```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import spearmanr

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
first_year_trend = pd.read_csv('/Users/kelvinfoo/Desktop/Crypto Research/Bitcoin and Ethereum Basic Statistics/first_year_trend.csv')
first_year_trend.head()
```

#### **Spearman Correlation**

```python
pivot_table = first_year_trend.pivot(index='Date', columns='currency', values='num_transactions')
spearman_corr = pivot_table.corr(method = 'spearman')

spearman_corr 
```

```python
pivot_table = first_year_trend.pivot(index='Date', columns='currency', values='total_value_usd')
spearman_corr = pivot_table.corr(method = 'spearman')

spearman_corr 
```

#### **ACF and PACF**

```python
btc_transaction = first_year_trend[first_year_trend['currency'] == 'BTC']['num_transactions']
eth_transaction = first_year_trend[first_year_trend['currency'] == 'ETH']['num_transactions']
doge_transaction = first_year_trend[first_year_trend['currency'] == 'Dogecoin']['num_transactions']

btc_value = first_year_trend[first_year_trend['currency'] == 'BTC']['total_value_usd']
eth_value = first_year_trend[first_year_trend['currency'] == 'ETH']['total_value_usd']
doge_value = first_year_trend[first_year_trend['currency'] == 'Dogecoin']['total_value_usd']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

plot_acf(btc_transaction, lags=90, ax=axs[0, 0])
axs[0, 0].set_title('ACF of number of transactions for BTC')

plot_acf(eth_transaction, lags=90, ax=axs[0, 1])
axs[0, 1].set_title('ACF of number of transactions for ETH')

plot_acf(doge_transaction, lags=90, ax=axs[1, 0])
axs[1, 0].set_title('ACF of number of transactions for Dogecoin')

fig.delaxes(axs[1, 1])
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

plot_pacf(btc_transaction, lags=50, ax=axs[0, 0])
axs[0, 0].set_title('PACF of number of transactions for BTC')

plot_pacf(eth_transaction, lags=50, ax=axs[0, 1])
axs[0, 1].set_title('PACF of number of transactions for ETH')

plot_pacf(doge_transaction, lags=50, ax=axs[1, 0])
axs[1, 0].set_title('PACF of number of transactions for Dogecoin')

fig.delaxes(axs[1, 1])
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

plot_acf(btc_value, lags=90, ax=axs[0, 0])
axs[0, 0].set_title('ACF of total value transacted (USD) for BTC')

plot_acf(eth_value, lags=90, ax=axs[0, 1])
axs[0, 1].set_title('ACF of total value transacted (USD) for ETH')

plot_acf(doge_value, lags=90, ax=axs[1, 0])
axs[1, 0].set_title('ACF of total value transacted (USD) for Dogecoin')

fig.delaxes(axs[1, 1])
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

plot_pacf(btc_value, lags=50, ax=axs[0, 0])
axs[0, 0].set_title('PACF of total value transacted (USD) for BTC')

plot_pacf(eth_value, lags=50, ax=axs[0, 1])
axs[0, 1].set_title('PACF of total value transacted (USD) for ETH')

plot_pacf(doge_value, lags=50, ax=axs[1, 0])
axs[1, 0].set_title('PACF of total value transacted (USD) for Dogecoin')

fig.delaxes(axs[1, 1])
plt.tight_layout()
plt.show()
```
